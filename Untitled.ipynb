```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
import matplotlib.ticker as mtick
%matplotlib inline
```

```python
telco_base_data=pd.read_csv('Churn_Modelling.csv')
```

```python
telco_base_data.head()
```

```python
telco_new=telco_base_data[['Geography', 'Gender', 'Exited']]
```

```python
telco_new.head()
```

## We use first count plot of seoborn to analyze the differnce between female and male churn by the countries

```python
sns.countplot(x='Geography', hue='Exited', data=telco_new)
plt.title('Churn Count By Geography')
plt.ylabel('Number Of Customers')
plt.xlabel('Countries')
plt.show()
```

The count plot indicates that churn is highest in Germany and lowest in Spain. France has the largest number of customers in the dataset.

## Now we use Hist Plot 

```python
sns.histplot(x='Gender', hue='Geography', data=telco_new, multiple='dodge')
plt.show()
```

```python
churn_df=telco_new[telco_new['Exited']==1]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(x='Gender',hue='Geography', data=churn_df, multiple='dodge')
plt.title('Gender Distribution of Churned Customers by Country')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(telco_base_data.Age[telco_base_data["Exited"]==0],color="red",fill=True,label="No Churn")
sns.kdeplot(telco_base_data.Age[telco_base_data["Exited"]==1],color="blue",fill=True,label="Churn")
plt.xlabel("Age")
plt.ylabel("Density")
plt.title("KDE Plot of Age: Churn vs No Churn")
plt.legend(loc="upper right")
plt.show()
```

```python
sns.kdeplot(telco_base_data.Balance [(telco_base_data["Exited"]==0)], color="Red" , fill=True)
sns.kdeplot(telco_base_data.Balance [(telco_base_data["Exited"]==1)], color="Blue" , fill=True)
plt.xlabel("Balance")
plt.ylabel("Density")
plt.title("KDE Plot of Balance: Churn vs No Churn")
plt.legend(loc="upper right")
plt.show()
```

```python
labels=['0-20', '21-40', '41-60', 'Above 60']
bins=[0, 20, 40, 60, 100]
telco_base_data['Age_bins']=pd.cut(telco_base_data.Age, bins, labels = labels , include_lowest=True)
```
